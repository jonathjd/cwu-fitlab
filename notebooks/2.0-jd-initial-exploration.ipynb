```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv('/Users/jdickinson/Documents/PersonalRepos/cwu-fitlab/data/processed/nhanes_merged.csv')
df.head(3)
```

```python
df_vo2 = df[df['CVDESVO2'].notna()]
df_vo2.head()
```

```python
men_df = df_vo2[df_vo2['RIAGENDR'] == 1]
fem_df = df_vo2[df_vo2['RIAGENDR'] == 2]
```

```python
fig = px.histogram(
    df_vo2, 
    x='CVDESVO2', 
    nbins=25,
    title="Histogram of VO2max",
    labels={
        "CVDESVO2": "Estimated VO2max (ml/kg/min)"
    },
    template="simple_white",
    color_discrete_map={
        "CVDESVO2": "red"
    }
)

line_placement = df_vo2['CVDESVO2'].mean()
fig.add_vline(
    type="line",
    line_color="salmon",
    line_width=4,
    opacity=1,
    line_dash="dot",
    x=line_placement
)

fig.show()
```
