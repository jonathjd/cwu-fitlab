```python
import pandas as pd
```

```python
# Script to load datasets and metadata into giant dataset
# Dexa data
dexa_00 = pd.read_sas('/Users/jdickinson/Documents/PersonalRepos/cwu-fitlab/data/raw/dxx00.xpt')
dexa_01 = pd.read_sas('/Users/jdickinson/Documents/PersonalRepos/cwu-fitlab/data/raw/dxx01.xpt')
dexa_03 = pd.read_sas('/Users/jdickinson/Documents/PersonalRepos/cwu-fitlab/data/raw/dxx03.xpt')
dexa_06 = pd.read_sas('/Users/jdickinson/Documents/PersonalRepos/cwu-fitlab/data/raw/dxx06.xpt')

# Metadata / demographic data
meta_00 = pd.read_sas('/Users/jdickinson/Documents/PersonalRepos/cwu-fitlab/data/raw/DEMO_00.XPT')
meta_02 = pd.read_sas('/Users/jdickinson/Documents/PersonalRepos/cwu-fitlab/data/raw/DEMO_02.XPT')
meta_04 = pd.read_sas('/Users/jdickinson/Documents/PersonalRepos/cwu-fitlab/data/raw/DEMO_04.XPT')

# verify
print(meta_00.shape)
print(meta_02.shape)
print(meta_04.shape)
print('\n')

# concatenate then merge
df_frames = [dexa_00, dexa_01, dexa_03, dexa_06]
meta_frames = [meta_00, meta_02, meta_04]
df_concat = pd.concat(df_frames)
meta_concat = pd.concat(meta_frames)
merged = df_concat.merge(right=meta_concat, how='inner', on='SEQN')

# verify
print(merged.shape)
```

```python
merged.to_csv('dexa_nhanes.csv', index=False)
```
